# 会员数据化运营

# 1 会员数据化运营概述

数据化运营可用来解决以下几方面问题：

    1.会员的生命周期状态是什么；
    
    2.会员的核心诉求是什么；
    
    3.会员的转化习惯和路劲是什么；
    
    4.会员的价值是几何；
    
    5.如何扩大市场覆盖，获得更多新会员；
    
    6.如何更好的维系老客户；
    
    7.应该在什么时间、采取何种措施、针对哪些会员做哪些运营活动；
    
    8.在特定运营目标下，应该如何制定会员管理策略，包括：行为管理、体验管理、增值服务、信息管理、营销管理、客户关怀等。

# 2 会员数据化运营关键指标

## 2.1 会员整体指标

    1.注册会员数
    
    2.激活会员数
    
    3.购买会员数

## 2.2 会员营销指标

    1.可营销会员数：手机号、邮箱、QQ、微信等
    
    2.营销费用：媒介费用、优惠券费用、积分兑换费用
    
    3.营销收入
    
    4.用券会员/金额/订单比例
    
    5.营销费率：会员营销费用占营销收入比例
    
    6.每注册/订单/会员收入
    
    7每注册/订单/会员成本

## 2.3 会员活跃度指标

    1.整体会员活跃度：定义会员活跃度矩阵，通过业务定义的关键因素判断整体会员活跃度
    
    2.每日/每周/每月活跃用户数

## 2.4 会员价值度指标

    1.会员价值分群：给用户打标签，显示用户状态、层次和价值区分
    
    2.复购率：一定周期内购买2次或2次以上的会员比例
    
    3.消费频次：分析消费黏性
    
    4.最近一次购买时间、金额

## 2.5 会员终生价值指标

    1.会员生命周期价值/订单量/平均订单价值：从用户成为企业会员到现在的总数据
    
    2.会员生命周期转化率：用户消费次数和到达网站/门店次数比例
    
    3.会员生命周期剩余价值：预测性指标

## 2.6 会员异动指标

    1.会员流失率：流失会员，使用拐点理论（Elbow Method）选择最合适的流失周期。
    
    2.会员异动比=新增购买会员/流失会员

# 3 会员数据化运营应用场景

    1.会员营销：
        1.建立客户会员管理体系，促进会员数据信息化
        2.将普通用户拓展为企业会员，并提高新会员留存率
        3.基于历史消费记录，挖掘用户潜在消费需求及消费热点
        4.基于历史数据，为会员营销活动提供策略指导和建议，促进精准营销活动的开展
        5.从会员营销结果中损招异常订单或转化，作为黄牛或VIP客户识别的参考
        6.挖掘会员传播关系，找到口碑传播效应的关键节点
    2.会员关怀:
        1.为预警时间设置阈值，自动触发应急处理机制
        2.分析会员行为，为会员提供个性化、精准化和差异化服务
        3.通过会员喜好分析，提高客户忠诚度、活跃度和黏性
        4.通过会员分析，预防会员流失并找到挽回已经流失会员的方法
        5.基于会员群体行为，更好的划分会员群体属性并挖掘群体性特征
        6.基于躯体用户和内容相似度，发现有价值的会员互动方式
        7.基于会员生命周期的关怀管理，促进用户终生价值最大化
    


# 4 会员数据化运营模型

    1.会员细分模型
        1.基于属性的方法：会员地域、产品类别、会员类别、会员性别、会员消费等级、会员等级
        2.ABC分类法：先将目标数据列倒序排序，然后做累计百分比统计。A（0%-80%），B（80%-90%），C(90%-100%)
        3.聚类法
    2.会员价值的模型
        RFM模型：R(Recency)、F(Frequency)、M(Monetary)
        计算RFM得分，通过三个维度来评估客户的订单活跃价值，常用来做客户分群和价值区分。
        思路1：基于三个维度指做用户群体划分和解读，对用户价值度做分析
                   例如得分为212的会员往往购买频率较低，针对购买频率低的客户定期发送促销活动邮件。
                   例如得分为321的会员虽然购买频率高但是订单金额低，考虑通过关联或搭配销售提升订单金额。
        思路2: 基于RFM总得分评估所有会员价制度价值，并排序。同时该得分可以作为输入维度和其他维度一起做数据挖掘的输入变量
    3.会员活跃度模型
        RFE模型：R(Recency)、F(Frequency)、E(Engagements)
    
    4.会员流失预测模型
        常见的属于流失的状态定义：
            1.会员已经退订公司的促销活动
            2.会员打电话要求将自己的信息加入通知黑名单
            3.会员已经连续6个月没有登录过网站
            4.针对会员发送的关怀激励活动没有任何有效反馈和互动
            5.会员最近一年内没有任何订单
        做会员流失预警模型时，需要注意的问题：
            1.流失会员的样本分类一定是少数类，需要处理样本不均衡问题。
            2.对于流失会员的预测结果，得到概率性的输出可以结合预测标签一起用
            3.对于参与训练模型的维度变量的选择要结合业务经验。
            4.会员流失预警模型不是一次性的，需要周期性的监视和运行
    5.会员特征分析模型
        应用场景：
            没有任何前期经验或特定目标下触发。（辅助，提示，启发性）
                1.聚类：分析不同群组的典型特征和群组间的差异，进行差异化管理
                2.统计分析：将整体用户做统计分析，了解整体数据概况
            有明确的欲望房型，希望知道能达到事件目标的会员特征，进行进一步会员运营（直接触点）
                1.分类：利用规则进而确定会员特征。
                2.关联：使用关联规则找到不同属性、项目间的关联发生或序列发生关系
                3.异常检测：使用非监督式的异常检测方法，从一堆数据中找到异常数据样本
    6.营销响应预测模型
        1.营销响应预测模型通常在做会员营销活动之前，通过分析找到可能响应活动的会员特征及整体响应用户比例、数量和可能带来的销售额
        2.营销响应预测模型一般采用分类算法，做预测模型前，首先需要收集训练所需的数据集
        3.预测模型的结果：
            1.基于模型找到最可能产生购买转化行为的会员规则特征
            2.基于模型预测可能产生的订单转化数量，转化率以及转化客户的客单价
