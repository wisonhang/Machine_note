```python
import pandas as pd 
import re
from random import randint
import re

def wordListSum(wordList):
    sum = 0
    for word, value in wordList.items():
        sum = sum + value
    return sum


def retrieveRandomWord(wordList):

    randomIndex = randint(1, wordListSum(wordList))
    for word, value in wordList.items():
        randomIndex -= value
        if randomIndex <= 0:
            return word



def buildWordDict(text):
    a=re.compile('F16.*')
    text=a.sub('F16',x)
    words = text.split('>')
    #print(text)
    words = [word for word in words if  word!=""]
    #print(words)

    for i in range(1, len(words)):
        if words[i-1] not in wordDict:
            wordDict[words[i-1]] = {}
        if words[i] not in wordRef:
            wordRef[words[i]]={}
        if words[i-1] not in wordRef[words[i]]:
            wordRef[words[i]][words[i-1]]=0
        if words[i] not in wordDict[words[i-1]]: 
            wordDict[words[i-1]][words[i]] = 0
        wordDict[words[i-1]][words[i]] = wordDict[words[i-1]][words[i]] + 1
        wordRef[words[i]][words[i-1]] = wordRef[words[i]][words[i-1]] + 1

    return(wordDict)

    
def filter_l(data,pct=0.01):
    #print(data.items())
    return {k: v for k, v in data.items() if (v/sum(data.values()))>pct}











text_data=pd.read_excel('Android Converted Tracks 20191014.xlsx')
text_list=list(text_data.track)


wordDict = {}
wordRef={}
for x in text_list:
    wordDict = buildWordDict(x)
wordRef={k:list(filter_l(v).keys()) for k,v in wordRef.items() if k!='F16'}

for i in wordRef:
    for j in wordRef[i]:
        if j in wordDict[i]:
            del wordDict[i][j]
###########删去ref路径转移点
# for i in wordRef:
#     for j in wordRef[i]:
#         if j in wordDict[i]:
#             del wordDict[i][j]
            
            
# def wordDict_generate(x):
#     wordDict = {}
#     wordRef={}
#     for tt in x:
#         buildWordDict(tt,wordDict,wordRef)
#         #print(wordDict)
#         #print('aa')
#         #print(wordRef)
#     wordRef={k:list(filter_l(v).keys()) for k,v in wordRef.items() if k!='F16'}
#     for i in wordRef:
#         for j in wordRef[i]:
#             if j in wordDict[i]:
#                 del wordDict[i][j]
#     return(wordDict,wordRef)
    
```

```python
def path_generate():
    length = 25
    path_list={}
    for _ in range(1,1000000):
        chain = ""
        currentWord = 'A00'
        for i in range(0, length):
            chain += currentWord + " "
            #print(chain)
            if wordDict[currentWord]:
                currentWord = retrieveRandomWord(wordDict[currentWord])
                if currentWord=='F16' or i ==length:
                    chain += currentWord + " "
                    if chain not in path_list:
                        path_list[chain]=1
                    else:
                        path_list[chain]+=1
                    break
            else:
                if chain not in path_list:
                    path_list[chain]=1
                else:
                    path_list[chain]+=1            
                break




    data=pd.DataFrame.from_dict(path_list,orient='index',columns=['count']).reset_index().sort_values('count',ascending=False)
    data=data.loc[data['count']>1000,:]
    data.index=range(0,len(data)) 
    path_set=list(data['index'])

    for A in path_set:
        def same_match(x,txt=A):
            return(x!=A and A.replace(' F16 ',' ') in x)
        temp=list(map(same_match,path_set))
        path_set=[path_set[i] for i ,j in enumerate(temp) if j==False ]
    data1=data.loc[data['index'].isin(path_set),]
    return(data,data1)
```
